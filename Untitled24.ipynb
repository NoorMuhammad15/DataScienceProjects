```python
#set chart style
sns.set(style="whitegrid")
#create figure from plt
plt.figure(figsize=(8,5))
#get the size of dataset
total = float(len(df))
#create chart ax with column sex and datasource equals to dataset
ax = sns.countplot(x="employee", data=df)
#calculate the percentage and put it on the top of bars with below code
for p in ax.patches:
    #how the percentage should be formated
    percentage = '{:.1f}%'.format(50 * p.get_height()/total)
    # % position in X
    x = p.get_x() + 0.4
    # % position in Y
    y = p.get_height()
    #put percentage on chart
    ax.annotate(percentage, (x, y),ha='center')
#plot the chart
plt.show()
```

```python
# using .get_dummies function to convert
# the categorical datatype to numerical
# and storing the returned dataFrame
# in a new variable df1
df1 = pd.get_dummies(df['companyAlias'])

# using pd.concat to concatenate the dataframes
# df and df1 and storing the concatenated
# dataFrame in df.
df = pd.concat([df, df1], axis=1).reindex(df.index)

# removing the column 'Purchased' from df
# as it is of no use now.
df.drop('companyAlias', axis=1, inplace=True)

# printing df
df
```
